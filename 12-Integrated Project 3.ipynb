# Data description

**Technological process**
- *Rougher feed* — raw material
- *Rougher additions*/*reagent additions* — flotation reagents: *Xanthate, Sulphate, Depressant*
    - *Xanthate* — promoter or flotation activator;
    - *Sulphate* — sodium sulphide for this particular process;
    - *Depressant* — sodium silicate.
- *Rougher process* — flotation
- *Rougher tails* — product residues
- *Float banks* — flotation unit
- *Cleaner process* — purification
- *Rougher Au* — rougher gold concentrate
- *Final Au* — final gold concentrate

**Stages' parameters**
- *air amount* — air volume
- *fluid levels*
- *feed size* — feed particle size
- *feed rate*

**Feature naming** `[stage].[parameter_type].[parameter_name]`

`[stage]`:
- *rougher —* flotation
- *primary_cleaner* — primary purification
- *secondary_cleaner* — secondary purification
- *final* — final characteristics

`[parameter_type]`:
- *input* — raw material parameters
- *output* — product parameters
- *state* — parameters characterizing the current state of the stage
- *calculation —* calculation characteristics

##  Check that recovery is calculated correctly. Using training set, calculate recovery for `rougher.output.recovery` feature. Find MAE between calculations and feature values. Provide findings.


- C — gold share in concentrate right after flotation (for finding rougher concentrate recovery)/after purification (for finding final concentrate recovery)

```python
#Saving columns as variables
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']

#Calculating recovery
train['recovery'] = (((C * (F - T)) / (F * (C - T))) * 100).round(6)

#Printing first five rows to compare
train[['recovery', 'rougher.output.recovery']].head()
```

## Write a function to calculate final sMAPE value.

sMAPE - symmetric Mean Absolute Percentage Error. Unlike MAE, it is expressed in relative values instead of absolute ones, equally taking into account the scale of both target and prediction.



Denotation:
- Target value for i index observation in sample used to measure quality.
- Prediction value for i index observation, for example, in test sample.
- Number of observations in sample.
- Summation over all sample observations (i takes values from 1 to N).

We need to predict two values:
1. rougher concentrate recovery `rougher.output.recovery`
2. final concentrate recovery `final.output.recovery`

Final metric includes two values:


```python
#Defining function to calculate sMAPE
def smape(y_true, y_pred):
    return (np.abs(y_true-y_pred)/((np.abs(y_true) + np.abs(y_pred))/2)).mean()

#Defining function to calculate final sMAPE
def final_smape(y_true,y_pred):
    smape_rougher = smape(y_true[:,0], y_pred[:,0])
    final_smape = smape(y_true[:,1], y_pred[:,1])
    return 0.25*smape_rougher + 0.75*final_smape
```

## Train different models. Evaluate them using cross-validation. Pick best model and test it using test sample. Provide findings.

### Model training on train set

```python
#Defining cross-validation function
def crossval(model, X_train, y_train):
    smape_score = make_scorer(final_smape)
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring=smape_score)
    return scores.mean()
```

```python
#Creating models
DR = DummyRegressor(strategy='mean')
LR = LinearRegression()
DTR = DecisionTreeRegressor(random_state=12345)
RFR = RandomForestRegressor(random_state=12345)

#Cross-validation to select best model
final_smape_DR = crossval(DR, X_train, y_train)
final_smape_LR = crossval(LR, X_train, y_train)
final_smape_DTR = crossval(DTR, X_train, y_train)
final_smape_RFR = crossval(RFR, X_train, y_train)

#Plotting
models = pd.DataFrame({
    'Model': ['Baseline', 'LR', 'DTR', 'RFR'],
    'Score': [final_smape_DR, final_smape_LR, final_smape_DTR, final_smape_RFR]})
sorted_scores = models.sort_values(by='Score', ascending=False)

ax=sns.barplot(x='Score', y = 'Model', data = sorted_scores)
ax.bar_label(ax.containers[0])
plt.title('sMAPE per model')
plt.xlabel('sMAPE')
plt.ylabel('Model');
```

### Hyperparameter tuning

```python
#Running loop to find best max_depth
d = []
for depth in range(1,11):
    DTR = DecisionTreeRegressor(random_state=12345, max_depth=depth)
    final_smape_DTR = crossval(DTR, X_train, y_train)
    d.append({'max_depth': depth, 'final_smape_DTR': final_smape_DTR})
        
best_param = pd.DataFrame(d).nsmallest(1, ['final_smape_DTR'], keep='first')
final_smape_DTR = best_param['final_smape_DTR'].values
best_param
```

```python
#Running loop to find best n_estimators and max_depth
d = []
for estim in range(1,51,9):
    for depth in range(1,11):
        RFR = RandomForestRegressor(random_state=12345, n_estimators=estim, max_depth=depth)
        final_smape_RFR = crossval(RFR, X_train, y_train)
        d.append({'n_estimators': estim, 'max_depth': depth, 'final_smape_RFR': final_smape_RFR})

best_param = pd.DataFrame(d).nsmallest(1, ['final_smape_RFR'], keep='first')
final_smape_RFR = best_param['final_smape_RFR'].values
best_param
```

RandomForestRegressor has better average score.

```python
#Splitting data into train and test sets (X- features, y-target)
X_train = train_params.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)
X_test =  test_params.drop(['rougher.output.recovery', 'final.output.recovery'], axis=1)
y_train = train_params[['rougher.output.recovery', 'final.output.recovery']].values
y_test = test_params[['rougher.output.recovery', 'final.output.recovery']].values

#Standard scaling
sc = ss()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```
